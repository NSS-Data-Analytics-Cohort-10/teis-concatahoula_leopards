```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re
```

```python
teis_df=pd.read_excel('../data/bdi3_all_evals.xlsx')
```

```python
teis_df=teis_df.drop_duplicates()
teis_df.reset_index()
teis_df= teis_df.rename(columns={'Location - Sub Level 1':'Location'})
```

```python
cog_sub_df=teis_df[['Child ID','Gender', 'Cognitive-Attention and Memory SS','Cognitive-Reasoning / Academic Skills SS','Cognitive-Perception and Concepts SS']]
cog_sub_df
```

```python
cog_am_avg = cog_sub_df.groupby('Gender')['Cognitive-Attention and Memory SS'].agg(np.mean).to_frame()
cog_am_avg
```

```python
cog_ra_avg = cog_sub_df.groupby('Gender')['Cognitive-Reasoning / Academic Skills SS'].agg(np.mean).to_frame()
cog_ra_avg
```

```python
cog_pc_avg = cog_sub_df.groupby('Gender')['Cognitive-Perception and Concepts SS'].agg(np.mean).to_frame()
cog_pc_avg
```

```python
cog_sub_avg= cog_am_avg.merge(cog_ra_avg, how='outer', on='Gender')
cog_sub_avg
```

```python
cog_sub_avg= cog_sub_avg.merge(cog_pc_avg, how='outer', on='Gender')
cog_sub_avg
```
